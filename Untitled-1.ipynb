```python
A = filtered_df[filtered_df['landing_page'] == '/lander-2']['clickthrough_rate_pct'].values
B = filtered_df[filtered_df['landing_page'] == '/lander-5']['clickthrough_rate_pct'].values

u_val, p_val = stats.mannwhitneyu(A, B, alternative='two-sided')

alpha = 0.05

print(f"--- MANN-WHITNEY U TEST RESULTS: /lander-2 vs /lander-5 ---")
print(f"Lander-2 Mean CTR: {np.mean(A):.2f}% (n={len(A)})")
print(f"Lander-5 Mean CTR: {np.mean(B):.2f}% (n={len(B)})")
print(f"Lander-2 Median CTR: {np.median(A):.2f}%")
print(f"Lander-5 Median CTR: {np.median(B):.2f}%")

print("-" * 40)
print(f"U-statistic:      {u_val:.4f}")
print(f"P-Value:          {p_val:.10f}")

print('\n--- P-value Decision ---')
if p_val < alpha:
    print('Reject H0. Significant difference found.')
    print('The distributions of the two groups are significantly different.')
else:
    print('Fail to reject H0. No strong evidence of difference.')
```

```python
A = filtered_df[filtered_df['billing_version'] == '/billing']['billing_to_success_cvr'].values
B = filtered_df[filtered_df['billing_version'] == '/billing-2']['billing_to_success_cvr'].values

u_val, p_val = stats.mannwhitneyu(A, B, alternative='two-sided')

alpha = 0.05

print(f"--- MANN-WHITNEY U TEST RESULTS: /billing vs /billing-2 ---")
print(f"Billing Mean CTR: {np.mean(A):.2f}% (n={len(A)})")
print(f"Billing-2 Mean CTR: {np.mean(B):.2f}% (n={len(B)})")
print(f"Billing Median CTR: {np.median(A):.2f}%")
print(f"Billing-2 Median CTR: {np.median(B):.2f}%")

print("-" * 40)
print(f"U-statistic:      {u_val:.4f}")
print(f"P-Value:          {p_val:.10f}")

mean_diff = np.mean(B) - np.mean(A)
df_approx = len(A) + len(B) - 2
crit_t = stats.t.ppf(1 - alpha/2, df_approx)
se = np.sqrt((np.var(A, ddof=1)/len(A)) + (np.var(B, ddof=1)/len(B)))
ci_lower = mean_diff - crit_t * se
ci_upper = mean_diff + crit_t * se

print(f"Mean Difference:   {mean_diff:.2f}%")
print(f"Approx. 95% CI:    [{ci_lower:.2f}%, {ci_upper:.2f}%]")

print('\n--- P-value Decision ---')
if p_val < alpha:
    print('Reject H0. Significant difference found.')
    print('The distributions of the two groups are significantly different.')
else:
    print('Fail to reject H0. No strong evidence of difference.')
```
